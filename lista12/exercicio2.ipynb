```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def arctan(z):
    return np.arctan(z)

def Leaky_relu(z):
    return np.maximum(0.1*z, z)

def softplus(z):
    return (np.log(1+np.exp(z)))

def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, arctan(z), "g--", linewidth=2, label="Arctan")
plt.plot(z, Leaky_relu(z), "b-", linewidth=2, label="Leaky relu")
plt.plot(z, softplus(z), "m-.", linewidth=2, label="Softplus")
plt.grid(True)
plt.legend(loc="best", fontsize=14)
plt.title("Activation functions", fontsize=14)
#plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(0, 0, "ro", markersize=5)
plt.plot(0, 0, "rx", markersize=10)
plt.plot(z, derivative(arctan, z), "g--", linewidth=2, label="Arctan")
plt.plot(z, derivative(Leaky_relu, z), "b-", linewidth=2, label="Leaky relu")
plt.plot(z, derivative(softplus, z), "m-.", linewidth=2, label="Softplus")
plt.grid(True)
plt.legend(loc="best", fontsize=14)
plt.title("Derivatives", fontsize=14)
#plt.axis([-5, 5, -0.2, 1.2])

plt.show()
```
