```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000000)

plt.subplot(1,2,1)
plt.title('PDF')
plt.hist(data , bins=100 , density=True , color='b')

plt.subplot(1,2,2)
plt.title('CDF')
plt.hist(data , bins=100, density=True , color='g' , cumulative=True)

plt.show()
plt.savefig('histograma.png')
```
