```python
import math
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-5,15,10000) 
listH = []

for i in x1:
    z = 5 - i
    h = 1/(1 + math.exp(-z))
    listH.append(h)

fig = plt.figure(figsize=(5, 5))
plt.fill_between(x1, 1, label = '- class')
plt.fill_between(x1, listH, label = '+ class')
plt.legend()
plt.show()
```
